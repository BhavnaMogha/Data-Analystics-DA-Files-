```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# multiple columns visualize
# perform aggregations in the graph
# Statistical analysis
```

```python
df = sns.load_dataset('tips')
df
```

# Numerical Columns
- relplot
- line plot
- scatterplot

```python
sns.relplot(x='total_bill',y='tip',data = df);
```

```python
sns.relplot(x='total_bill',y='tip',hue='day',data = df);
```

```python
sns.relplot(x='total_bill',y='tip',hue='day',size='sex',data = df);
```

```python
df['size'].unique()
```

```python
sns.relplot(x='total_bill',y='tip',hue='day',size='sex',style='size',data = df);
```

```python
sns.relplot(x='total_bill',y='tip',data = df,hue='day',size='sex',style='size',col='time',row='smoker');
```

```python
sns.lineplot(x='total_bill',y='tip',data = df);
```

```python
sns.scatterplot(x='total_bill',y='tip',data = df);
```

```python
# Categorical Columns

# Catplot-----strip plot
# swarm plot
# barplot
# histplot
# count plot
# boxplot
```

```python
sns.catplot(x='day',y='total_bill',hue='day',data = df); # strip plot
```

```python
sns.catplot(x='day',y='total_bill',hue='sex',kind='swarm',data = df); # strip plot
```

```python
d = df.groupby(['day'])['total_bill'].agg(['sum','mean','max','min']).reset_index()
d
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.barplot(x='day',y='total_bill',estimator='median',data=df,ci=None,palette='Set1');
```

```python
sns.catplot(x='day',y='total_bill',estimator='median',kind='bar',data=df,ci=None,palette='Set1');
```

```python
sns.histplot(x='size',data=df);
```
