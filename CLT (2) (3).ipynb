```python
import pandas as pd
```

```python
df1=pd.read_csv('train (2).csv')
df1
```

```python
df1.drop(columns=['Survived'],inplace=True)
```

```python
df2=pd.read_csv('test (2).csv')
df2
```

```python
df=pd.concat([df1,df2],axis=0)
df
```

```python
df.sample(1309)
```

```python
# df.sample(1309)
```

```python
# shuffling the data
df=df.sample(df.shape[0])
df
```

```python
# df['Fare'].mean()# popluation
```

```python
df['Fare'].isnull().sum()
```

```python
fare=df['Fare'].dropna(axis=0)
```

```python
fare
```

```python
fare=pd.Series(fare)
fare
```

```python

for i in range(3):
    a=fare.sample(2)
    print(a)
```

```python
# collecting the 100 samples and the sample size is 40

sample=[]
for i in range(100):
    sample.append(fare.sample(40).values.tolist())
```

```python
sample=np.array(sample)
sample
```

```python
sample.shape
```

```python
samp_dist=np.mean(sample,axis=1)# sampling distribution
samp_dist
```

```python
sns.distplot(df['Fare']);
```

```python
sns.distplot(samp_dist);
```

```python
np.mean(samp_dist)# sample mean
```
