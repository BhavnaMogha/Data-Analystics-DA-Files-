```python
# line plot - we can visualize the trend  ( it i want to show the analysi with respect to time interval)
```

```python
a = [9,1,5]
b = [2,7,9]

plt.plot(a,b) # here plot means line
plt.show()
```

```python
a = [9,1,5]
b = [2,7,5]

plt.plot(a,b,color='r',marker='D',markersize=10,linestyle='--',linewidth=3) # here plot means line
plt.title('Representing package w.r.t cgpa')
plt.xlabel('CGPA')
plt.ylabel('package')
plt.show()
```

```python
x = np.arange(1,11)
y =np.arange(11,21)

plt.figure(figsize=(5,2))
plt.plot(x,y,color='r',marker='o',markersize=10,linestyle='--',linewidth=3)
plt.title('linear relation')
plt.xlabel('x axis')
plt.ylabel('y axis')
print(x)
print(y)
plt.xticks(x)
plt.yticks(y);
# plt.show()
```

```python
plt.plot(x,x**2,color='r',marker='o',markersize=10,linestyle='--',linewidth=3);
```

```python
mobile=['samsung','oppo','Apple','lgg','mi']
online = np.random.randint(10000,20000,5)
offline = np.random.randint(10000,20000,5)

plt.plot(mobile,online,color='y',marker='o',markersize=4,linestyle='--',linewidth=1.5,label='online')
plt.plot(mobile,offline,color='k',marker='*',markersize=4,linestyle='--',linewidth=1.5,label='offline')
plt.title('Onlive Vs Oflline Price')
plt.xticks(rotation='vertical')
plt.xlabel('Brands')
plt.ylabel('Prices')
plt.legend();
```

```python
plt.figure(figsize=(12,4))

plt.subplot(1,2,1)
plt.plot(mobile,online,color='y',marker='o',markersize=4,linestyle='--',linewidth=1.5,label='online')
plt.legend()


plt.subplot(1,2,2)
plt.plot(mobile,offline,color='k',marker='*',markersize=4,linestyle='--',linewidth=1.5,label='offline')
plt.legend();

```

```python
plt.figure(figsize=(6,9))

plt.subplot(211)
plt.plot(mobile,online,color='y',marker='o',markersize=4,linestyle='--',linewidth=1.5,label='online')
plt.legend()


plt.subplot(212)
plt.plot(mobile,offline,color='k',marker='*',markersize=4,linestyle='--',linewidth=1.5,label='offline')
plt.legend();

```
