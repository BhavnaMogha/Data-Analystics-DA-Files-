```python
# 0quantile --------0perectile
# 1quantile --------25perectile
# 2quantile --------50perectile
# 3quantile --------75perectile
# 4quantile --------100perectile
```

```python
# upper_limit = q3+ 1.5 *(IQR)
# lower_limit= q1-1.5*(IQR)

# IQR = q3-q1
```

```python
df1['tip'].describe()
```

```python
q1 = np.percentile(df1['tip'],25)
q1
```

```python
q3= df1['tip'].quantile(0.75)
q3
```

```python
IQR=q3-q1
IQR
```

```python
u = q3+ 1.5 *(IQR)
u
```

```python
l = q1 - 1.5 *(IQR)
l
```

```python
df1[df1['tip']>u]
```

```python
df1[df1['tip']<l]
```

# Handling Outliers
- Trimming
- Capping (Winsorization)

```python
df1
```

```python
df1[df1['tip']<u]
```

```python
# Capping
```

```python
df1['tip2']=np.where(df1['tip']>u, u ,
         np.where(df1['tip']<l,l ,
                  df1['tip']))
```
